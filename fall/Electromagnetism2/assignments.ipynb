Three point charges lie on a line with neares-neighbour separation d as shown in the figure. The

* A) and b)

We know the formular for the force is $F(d_{12}) = k\frac{q_1q_2}{d_{12}^2}$, and as we know that -q pull on +Q is the same as the push between the two Q's. We have that $k\frac{qQ}{d^2} = k\frac{QQ}{(2d)^2}$

```python
import re
from sympy import Eq, solve, symbols


q,Q,k,d_Cu = symbols('q Q k d', real=True)
eq = Eq(k * q * Q / d_Cu**2, k * Q * Q / (2 * d_Cu)**2)
solve(eq, Q)
```

* C) and D)

For the point where the total electrostatic potential $V(x) = k\frac{Q}{r}$ is zero, we can use the fact that Q = 4q.

```python
from sympy import simplify


Q = 4*q
x = symbols('x', real=True)
V1 = k * Q / abs(x - 0)
V2 = k * -q / abs(x - d_Cu)
V3 = k * Q / abs(x - 2*d_Cu)
V_total = V1 + V2 + V3

# We check wiht x = d/2
print(f"The value of V_total at x = d/2 is: {V_total.subs(x, d_Cu/2).simplify()}")

# We check with x = 4d/5
print(f"The value of V_total at x = 4d/5 is: {V_total.subs(x, 4*d_Cu/5).simplify()}")
```

* E)
1.  **Region $x < 0$ (left of all charges):**
    * As $x \to -\infty$, all charges are far away, so $V(x) \to 0$.
    * As $x \to 0$ (from the left), you get infinitely close to the **positive** charge $+Q$. So, $V(x) \to +\infty$.
    * Since the function comes from $0$ and goes to $+\infty$, it must stay positive. It **never crosses zero** in this region.

2.  **Region $0 < x < d$ (between $+Q$ and $-q$):**
    * As $x \to 0$ (from the right), you are near the **positive** charge $+Q$. So, $V(x) \to +\infty$.
    * As $x \to d$ (from the left), you are near the **negative** charge $-q$. So, $V(x) \to -\infty$.
    * Because the potential is a continuous function in this interval, and it *must* go from $+\infty$ to $-\infty$, it **must cross $V=0$ exactly once** somewhere in this region. (This is a rule from calculus called the Intermediate Value Theorem).

3.  **Region $d < x < 2d$ (between $-q$ and $+Q$):**
    * As $x \to d$ (from the right), you are near the **negative** charge $-q$. So, $V(x) \to -\infty$.
    * As $x \to 2d$ (from the left), you are near the **positive** charge $+Q$. So, $V(x) \to +\infty$.
    * Just like in the previous region, the function is continuous and goes from $-\infty$ to $+\infty$. Therefore, it **must cross $V=0$ exactly once** in this region, too.

4.  **Region $x > 2d$ (right of all charges):**
    * As $x \to 2d$ (from the right), you are near the **positive** charge $+Q$. So, $V(x) \to +\infty$.
    * As $x \to +\infty$, all charges are far away, so $V(x) \to 0$.
    * Like in the first region, the function starts at $+\infty$ and approaches $0$ (from the positive side), so it must stay positive. It **never crosses zero** in this region.

Therefore, statement E) is **correct**.

We know that the electric field is the negative derivative (slope) of the potential: $E=dV/dx$, meaning we are looking for a point where the slope is zero, not where one axis is zero. At the two points we found where V=0, the slope dV/dx is not zero, meaning $E \neq 0$ and a test charge will experience a force.

The below table gives mass density, $\rho m$ molar mass, M, and number of conduction electrons per
atom, Z for the metals Ag, Cu, Al (silver, copper, aluminium).



a) Derive an expression for the drift velocity of the charge carriers, $v_d$ in a cable having a circular
cross section, as a function of the current I, and the cable (conductor) diameter d. Plot the drift
velocity as a function of d for I=10.0 A, in the three materials, e.g. for d between 0.1 and 100
mm.

b) Determine the resistance per length as a function of d in the three materials. Data for
resistivity can be found in the textbook, Table 9.1.

c) Determine the mass per meter for a cable with resistance 0.1 $\Omega/km$ for the three materials.

#### Answer
a)

```python
import numpy as np

I = Q_(10.0, 'A') # current
d_Cu = np.linspace(0.1, 100, 1000) * ureg.mm # diameter range

Ag_Z = 1; Ag_pm = Q_(1.05e4, 'kg/m^3'); Ag_M = Q_(0.108,  'kg/mole')
Cu_Z = 1; Cu_pm = Q_(8.80e3, 'kg/m^3'); Cu_M = Q_(0.0635, 'kg/mole')  
Al_Z = 3; Al_pm = Q_(2.70e3, 'kg/m^3'); Al_M = Q_(0.0270, 'kg/mole')

def drift_velocity(I, d, Z, pm, M):
    A = np.pi * (d/2)**2 # cross-sectional area
    
    # n = number density of charge carriers (electrons)
    # n = (atoms/m^3) * (conduction electrons/atom)
    n = (pm / M) * avogadro_number * Z 
    
    # q = charge of a single charge carrier (one electron)
    q = q_electron 
    
    v_d = I / (n * q * A)
    return v_d.to('mm/s')

v_d_ag = [drift_velocity(I, di, Ag_Z, Ag_pm, Ag_M).magnitude for di in d_Cu]
v_d_cu = [drift_velocity(I, di, Cu_Z, Cu_pm, Cu_M).magnitude for di in d_Cu]
v_d_al = [drift_velocity(I, di, Al_Z, Al_pm, Al_M).magnitude for di in d_Cu]

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(d_Cu.magnitude, v_d_ag, label='Silver', color='silver')
plt.plot(d_Cu.magnitude, v_d_cu, label='Copper', color='orange')
plt.plot(d_Cu.magnitude, v_d_al, label='Aluminum', color='blue')
plt.xscale('log')
plt.xlabel('Diameter (mm)')
plt.ylabel('Drift Velocity (mm/s)')
plt.title('Drift Velocity vs Diameter for Different Materials (I=10A)')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

Three point charges whose values are -4 nC, 0.5 nC and 2 nC are distributed on an equilateral
triangle with side length  as shown in the figure.
How are the values of the total electrical potential $_1$, $_2$ og $_3$, as measured in the points

#### Answer
Electrostatic potential (V) is a scalar quantity, not a vector. This means we don't worry about directions. The total potential at a point is simply the algebraic sum of the potentials from each individual charge.
$$
V = k\frac{q}{r}
$$


Let's label the charges:
* $q_A = -4$ nC (Top)
* $q_B = 0.5$ nC (Bottom-left)
* $q_C = 2$ nC (Bottom-right)

The charges form an equilateral triangle with side length $a$.
The points $V_1, V_2, V_3$ are located such that:
* $V_1$ forms an equilateral triangle with $q_A$ and $q_B$.
* $V_2$ forms an equilateral triangle with $q_A$ and $q_C$.
* $V_3$ forms an equilateral triangle with $q_B$ and $q_C$.

This means there are two distances we care about:
1.  **Short distance:** $a$
2.  **Long distance:** The distance from a vertex (like $V_1$) to the one charge it's *not* close to (like $q_C$). This distance is 2 times the height of a triangle with equal lenght sides. 
    * $h^2 + (a/2)^2 = a^2$
    * $h^2 = a^2 - (a/2)^2$
    * $h = \sqrt{a^2 - a^2/4} = \sqrt{3a^2/4} = \mathbf{a\sqrt{3}/2}$
    * So, the long distance is $2 \cdot h = 2 \cdot (a\sqrt{3}/2) = \mathbf{a\sqrt{3}}$.

We can now calculate the values


```python
from sympy import sqrt


a = symbols('a', real=True)
V1 = (-4 / a) + (0.5 / a) + (2 / (a * sqrt(3)))
V2 = (-4 / a) + (2 / a) + (0.5 / (a * sqrt(3)))
V3 = (0.5 / a) + (2 / a) + (-4 / (a * sqrt(3)))

print(f"Electrostatic potential at V1: {V1.evalf()}")
print(f"Electrostatic potential at V2: {V2.evalf()}")
print(f"Electrostatic potential at V3: {V3.evalf()}")
```
