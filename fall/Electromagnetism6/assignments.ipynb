## a) 
Determine the flux of the magnetic field through the solenoid as function of the rotation angle
, with the sign given by the normal vector

#### Answer
We know that the magnetic flux is defined as:
$$
\phi = \int B \cos{\theta} dA
$$
We also know that the magnetic field is homongenous (constant) and that the area is that of a circle ($\pi r^2$)
$$
\phi = B \cos{\theta} A = \pi r^2 B \cos{\theta}
$$


## b) 
Determine the induced emf and the current in the solenoid as function of time. Assume that
magnetic fields generated by current in the solenoid (self-induction) can be neglected in comparison with the external B-felt. Also assume that the resistance of the solenoid itself can be neglected.




#### Answer
To answer this we will use fridays law
$$
\varepsilon = -N\frac{d\phi}{dt}
$$
We already know the magnetic flux from a), and as the angle is the only thing that changes we need to look into the change in angle over time, we also need to remember to differentiate before we can move ($\pi r^2 B \cos{\theta}$) out, so we get
$$
\frac{d\Phi_B}{dt} = \frac{d}{dt} \left( \pi r^2 B \cos(\theta) \right)
$$
$$
\frac{d\Phi_B}{dt} = \pi r^2 B \cdot (-\omega \sin(\theta)) = -\pi r^2 B \omega \sin(\theta)
$$

We also know that the solenoid rotates with a constant angular speed $\omega$. Since the rotation starts at t=0 with $\theta$=0, we can express the angle as a function of time:
$$
\theta(t) = \omega t
$$
so
$$
\varepsilon = -1 (-\pi r^2 B \omega \sin(\theta)) = \pi r^2 B \omega \sin(\omega t)
$$

Now we just need to determine the current via ohms law
$$
I(t) = \frac{\varepsilon(t)}{R}
$$
we can now insert
$$
I(t) = \frac{\pi r^2 B \omega \sin(\omega t)}{R_0}
$$

## d)
Determine the torque on the solenoid caused by the magnetic field and the induced current. Be
careful with the sign  in which direction will the torque act on the solenoid

#### Answer
We know that the torque is defined by
$$
\vec{\tau} = \vec{\mu_0} \times \vec{B}
$$
where the magnetic dipole moment is
$$
\vec{\mu_0} = N I A \vec{n}
$$

So what we need to do is find out the direction in which the torque will point, but to do that we first need to find the direction of the dipole moment. Using Lenz's Law, which states that the induced effect always opposes the cause that produced it. In this case the change we introduce is the rotation of the shaft with the angular velocity ($\omega$). The effect is the induced current and the resulting magnetic torque.

Therefore, the magnetic torque always acts in the opposite direction to the mechanical rotation. It acts as a "braking" torque, trying to stop the solenoid from spinning. 

We can now find the size of the magnetic dipole to be
$$
\mu_0 = N I A = \frac{\pi r^2 B \omega \sin(\omega t)}{R_0} \cdot \pi r^2 = \frac{\pi^2 r^4 B \omega \sin(\omega t)}{R_0}
$$
now we can calculate the torque to be
$$
\tau(\theta) = \mu_0 B \sin{\theta}
$$
do we then change the formula to use the speed and time instead of angle we get
$$
\tau(t) =  \mu_0 B \sin{\omega t} = \frac{\pi^2 r^4 B^2 \omega \sin(\omega t)^2}{R_0}
$$


## e) 
If the angular speed should be kept constant, it implies that the torque determined in d) should
be counteracted by the torque from the shaft, i.e., this must be varying. A wind turbine will
naturally deliver a constant torque. Discuss how this problem can be solved or diminished.

#### Answer
The most effective way to eliminate the pulsation is to redesign the generator to use more than one phase, so that we instead use e.g. a Three-Phase (or generally polyphase) rather than single-phase.

Instead of a single solenoid loop, the generator uses three separate coils arranged 120 apart around the shaft. Each coil produces a pulsating torque similar to the one found in d), but they peak at different times. When you mathematically sum the torque from three coils offset by 120, the sine-squared terms add up to a constant number.

$$
\sin{\theta}^2 + \sin{\theta - 120}^2 + \sin{\theta - 240}^2 = \frac{3}{2}
$$
So the total resistive torque becomes constant. This matches the constant push of the wind, allowing for perfectly smooth rotation without vibration.

## f) 
Show that the resulting emf and current in the solenoid and the torque on the solenoid become
the same (show it following the method in question a)-e) but with reference to the situation
skethed in the figure below where the magnets and not the solenoid are turning). Discuss what
the practical implications are if the magnets are rotating in stead of the solenoid.

#### Answer
To show the results are the same, we simply change our reference frame.

- Solenoid (Fixed): The normal vector n of the solenoid area A is now fixed in space (e.g., pointing along the x-axis).
- Magnets (Rotating): The magnetic field vector B now rotates with angular velocity $\omega$.
- Angle: The angle $\theta$ between the fixed vector n and the rotating vector B changes with time: $\theta(t)=\omega t$.

The flux is the dot product of the field and the area vector 
$$
\Phi_B(t) = \vec{B}(t) \cdot \vec{A} = B A \cos(\theta) = B (\pi r^2) \cos(\omega t)
$$
*Result: This is the exact same formula as in Part (a) and therefore the other calculations will also be.*

While the magnitude is the same, the *effect* is slightly different. In the previous case, the torque tried to stop the spinning solenoid. In this case, the torque tries to twist the stationary solenoid in the direction the magnets are spinning. The *reaction* torque (on the spinning magnets) is what acts as the braking force, conserving energy. The smart thing is that by keeping the solenoid stationary, the high-power cables can be hard-wired directly to the transformer/grid. This is safer, cheaper, and more reliable.
