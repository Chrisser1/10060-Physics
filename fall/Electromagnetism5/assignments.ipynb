# Exercise 2


A proton having a magnetic dipole moment
$\mu_p=1.41 \cdot 10^{-26}$ J/T is situated in the middle of a long
coil, of which a section is shown in the figure, cut through the middle in the longitudinal
direction. The coil consists of a wires such that it has 500 turns per unit length. A current I=12.0
A is run through the coil in the direction shown in the figure. What is the difference in magnetic
potential energy between the two orientations of the proton dipole, A og B, shown in the figure?

A) $E_A - E_B = 2.13 \cdot 10^{-28} J$

B) $E_A - E_B = 1.06 \cdot 10^{-28} J$

C) $E_A - E_B = 5.32 \cdot 10^{-29} J$

D) $E_A - E_B = 3.38 \cdot 10^{-29} J$

E) $E_A - E_B = 0 J$

F) $E_A - E_B = -3.38 \cdot 10^{-29} J$

G) $E_A - E_B = -5.32 \cdot 10^{-29} J$

H) $E_A - E_B = -1.06 \cdot 10^{-28} J$

I) $E_A - E_B = -2.13 \cdot 10^{-28} J$

J) Dont know

#### Answer
A particle with a magnetic dipole momentum ($\mu$), like the proton, has potential energy (E) when it's in a magnetic field. This energy depends on its orientation relative to the field.
$$
U = -\mu \cdot \vec{B} = -\mu B cos(\theta)
$$
where $\theta$ is the angle between the momentum and the magnetic field B. What we need to do is find the magnetic field that the coil generates. We know that we are working with a coil, more specifically a solenoid coil, so:
$$
B = \mu_0 n I
$$
where n is the turns per unit length.

We also know that the magnetic field points to the left based on the right hand rule. Using this we can say that
$$
E_A = -\mu B cos(180) = \mu B
$$
and
$$
E_B = -\mu B cos(0) = -\mu B
$$

```python
mu_p = Q_(1.41e-26, 'J/T')
I = Q_(12, 'A')
n = Q_(500, 'm^-1')

B = mu_0 * n * I
print(f"Magnetic field strength (B): {B:.2e}")

E_A = mu_p * B
E_B = -mu_p * B
U = E_A - E_B
print(f"Potential energy (U): {U:.2e}") 
```

# Exercise 3




A circular conductor having a radius a = 20.0 cm is situated in a homogeneous magnetic field
pointing out of the plane of the figure. The total resistance of the conductor is $R_0=10.0$ W.

a) Determine the direction of the induced current if the magnetic field strength increases or
decreases.

b) Assume that the magnetic field is 0 for t<0, and grows as $B(t) = _0 (1  \exp{\frac{-t}{t0}})$ at later times. Derive an expression for the current in the conductor as a function of time, and plot the function
for $B_0=0.1$ T, $t_0=2.00$ ms.

#### Answer
a)

Based on Faradays law we know that if we have a change in the magnetic field we will generate a electromotive force that tries to reverse the change. So if we decrease it the current will run counterclockwise as that would generate a magnetic force outwards and on an increase the other way around.

b)
We know that the current in the conductor can be written as:
$$
I = \frac{\varepsilon(t)}{R}
$$
We also know that
$$
\varepsilon = |\frac{d\Phi_m}{dt}|
$$
Since $\vec{B}$ is perpendicular to the plane of the coil, the magnetic flux is given by
$$
\Phi_m = B \pi r^2 = _0 (1  \exp{\frac{-t}{t_0}}) \cdot \pi \cdot r^2
$$
Based on this we get:
$$
\varepsilon = |\frac{d}{dt} _0 (1  \exp{\frac{-t}{t_0}}) \cdot \pi \cdot r^2|
$$

```python
import sympy as sp
B0, t0, t, r = symbols('B0 t0 t r')
eq = sp.diff(sp.pi * r**2 * B0 * (1 - sp.exp(-t / t0)), t)
sp.latex(eq)
```

$$
\varepsilon = |\frac{\pi B_{0} r^{2} e^{- \frac{t}{t_{0}}}}{t_{0}}|
$$
now we have
$$
I = \frac{|\frac{\pi B_{0} r^{2} e^{- \frac{t}{t_{0}}}}{t_{0}}|}{R_0} = \frac{|\pi B_{0} r^{2} e^{- \frac{t}{t_{0}}}|}{R_0 t_{0}} 
$$

```python
import matplotlib.pyplot as plt
r = Q_(20, 'cm').to('m').m
t0 = Q_(2, 'ms').to('s').m
R0 = Q_(10, 'ohm').to('ohm').m
B0 = Q_(0.1, 'T').m

def I(t):
    return np.pi * B0 * (r)**2 * np.exp(-t / t0) / (R0 * t0)

t = np.linspace(0, 0.01, 5000)  # 0 to 10 ms

plt.figure()
plt.plot(t, I(t))
plt.xlabel("Time (s)")
plt.ylabel("Current I(t) (A)")
plt.title("Current as a Function of Time")
plt.grid(True)
plt.show()
```
