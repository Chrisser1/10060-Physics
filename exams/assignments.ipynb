### Answer
For at finde punktet hvor bilen kommer til at stoppe skal vi finde ud af hvornår bilen stopper kan vi bruge:
$$

# Spørgsmål 6.
En skihopper med massen 95.3 kg inkl. ski og påklædning, starter sit spring i hvile på toppen af en
skihopbakke, og lander 130 højdemeter nede med en fart på 28.0 m/s ved nedslaget. Hvor meget
varmeenergi, $E_v$ er blevet genereret af luftmodstand og glidefriktion i skihopperens bevægelse frem
til nedslagsøjeblikket? Tyngdeaccelerationen kan sættes til g=9.81 m/s2 på hopbakken.

A) $E_v = 37.4$ kJ

B) $E_v = 54.6$ kJ

C) $E_v = 84.2$ kJ

D) $E_v = 98.7$ kJ

E) $E_v = 121.$ kJ

F) $E_v = 159.$ kJ

G) $E_v = 178.$ kJ

H) Ved ikke

#### Answer
For at udregne dette bruger vi energi bevarelse. Vi ved at ved overfladen hvor skihopperen starter har personen en potentiel energi på
$$
U_{pot} = m g h
$$
Vi kan nu antage at skihopperen ved bunden har en kinetisk energi på
$$
U_{kin} = \frac{1}{2} m v^2
$$
Vi kan nu bruge dette til at finde varmetabet, da forskellen på de to er den
$$
U_{tot} = U_{pot} - U_{kin} 
$$

```python
m = Q_(95.3, 'kg')
g = Q_(9.81, 'm/s^2')
h = Q_(130, 'm')
v = Q_(28.0, 'm/s')

U_pot = m * g * h
U_kin = 0.5 * m * v**2
U_total = U_pot + U_kin
print(f"Total heat energy lost: {U_total.to('kJ'):.0f}")
```

# Spørgsmål 8.
En skiløber bevæger sig med konstant fart ned ad en bakke med konstant hældning på 21.3 grader i
forhold til vandret. Skiløberens masse inkl. ski og øvrigt udstyr er 100. kg. Den kinematiske
friktionskoefficient mellem sne og ski er 0.100. Tyngdeaccelerationen kan sættes til g=9.81 m/s2.

Hvad er størrelsen af den kraft $F_l$ som luftmodstanden udøver på skiløberen?

A) Fl =113. N

B) Fl =118. N

C) Fl =258. N

D) Fl =265. N

E) Fl =334. N

F) Fl =345. N

G) Fl =701. N

H) Fl =723. N

I) Fl =883. N

J) Fl =903. N

K) Ved ikke

#### Answer
For at løse dette kan vi bruge den viden vi har at der er bevægelse med konstant hastighed. Vi skal altså bare udregne den kraft der trækker skiløberen nedaf. 
I dette tilfælde er det kun tyngdekraften so gør dette. Vi skal dog huske at trække gnidningskraften fra. 
$$
f_g = m g \sin{\phi} - F_k
$$
hvor
$$
F_k = \mu_k F_N
$$
hvor $F_N$ er normalkraften
$$
F_N = m g \cos{\phi}
$$

```python
m = Q_(100, 'kg')
g = Q_(9.81, 'm/s^2')
mu = Q_(21.3, 'degrees').to('radians').magnitude
mu_k = 0.1

F_N = m * g * np.cos(mu)
F_K = mu_k * F_N
F_g = m * g * np.sin(mu) - F_K
print(f"Force: {F_g.to('N'):.1f}")
```
